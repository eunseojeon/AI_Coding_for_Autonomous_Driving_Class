##âœ… whileë¬¸ì„ ì´ìš©í•œ ììœ¨ì£¼í–‰ ì˜ˆì œ

ğŸš— 1. ì „ë°© ê±°ë¦¬ ì¤„ì–´ë“¤ ë•Œê¹Œì§€ ë°˜ë³µ ê°ì†

```python
distance = 50  # ì´ˆê¸° ì „ë°© ê±°ë¦¬

while distance > 30:
    print("ê°ì† ì¤‘... (ê±°ë¦¬:", distance, "m)")
    distance -= 5  # ì ì  ê°€ê¹Œì›Œì§

print("ê±°ë¦¬ ì¶©ë¶„íˆ ê°€ê¹Œì›€ â†’ ìœ ì§€ ì£¼í–‰")
```

ğŸš— 2. ì°¨ì„ ì´ ì¸ì‹ë  ë•Œê¹Œì§€ ê¸°ë‹¤ë¦¬ê¸°

```python
lane_detected = False
tries = 0

while not lane_detected and tries < 5:  # 5ë²ˆê¹Œì§€ë§Œ ì‹œë„
    print("ì°¨ì„  ì¸ì‹ ì‹œë„ ì¤‘...")
    tries += 1
    if tries == 3:  # 3ë²ˆì§¸ì— ì¸ì‹ëë‹¤ê³  ê°€ì •
        lane_detected = True

print("ì°¨ì„  ì¸ì‹ ì„±ê³µ!" if lane_detected else "ì°¨ì„  ì¸ì‹ ì‹¤íŒ¨")
```

ğŸš— 3. ì‹ í˜¸ë“±ì´ ì´ˆë¡ë¶ˆì´ ë  ë•Œê¹Œì§€ ì •ì§€

```python
signal = "red"
wait = 0

while signal != "green":
    print("ì‹ í˜¸ ëŒ€ê¸° ì¤‘... í˜„ì¬:", signal)
    wait += 1
    if wait == 3:
        signal = "green"  # 3ì´ˆ í›„ ì´ˆë¡ë¶ˆ ëë‹¤ê³  ê°€ì •

print("ì´ˆë¡ë¶ˆ! ì¶œë°œ!")
```

```python
offset = -0.5  # ì™¼ìª½ìœ¼ë¡œ ì¹˜ìš°ì³ ìˆìŒ

while offset < -0.1:
    print("ìš°íšŒì „ ì¡°í–¥... (offset:", offset, ")")
    offset += 0.1  # ì ì  ì¤‘ì•™ìœ¼ë¡œ ì´ë™

print("ì°¨ì„  ì¤‘ì•™ ë³µê·€ ì™„ë£Œ")
```
