```python
# ultralytics 라이브러리(최신 YOLO 모델 제공) 최신 버전 설치
!pip install -U ultralytics

from ultralytics import YOLO            # YOLO 모델 클래스 불러오기
from google.colab import files          # Colab에서 파일 업로드/다운로드 기능 사용
import glob                             # 파일 목록 검색용 모듈
import os                               # 파일 경로 조작용 모듈

# 1. YOLO12 공식 모델 로드  (yolo12n.pt 파일이 ultralytics에서 자동 다운로드된다고 가정)
model = YOLO("yolo12n.pt")              # YOLO12 nano(가정) 가중치로 모델 생성
model.info()                            # 모델 구조 및 파라미터 정보 출력

# 2. 분석할 동영상 파일 업로드 (Colab 환경에서 내 PC → Colab으로 이동)
uploaded = files.upload()               # 사용자에게 파일 업로드 창 제공
video_path = list(uploaded.keys())[0]   # 업로드한 첫번째 파일의 파일명(경로) 저장

# 3. 동영상을 대상으로 YOLO12 모델로 객체 인식 실행
results = model.predict(video_path, save=True)  # 결과(예: 바운딩 박스 표시 영상) 자동 저장

# 4. 저장된 결과 동영상 파일(.mp4) 경로 지정
result_dir = "runs/detect/predict"      # YOLO 추론 결과가 저장되는 폴더 경로
output_files = glob.glob(os.path.join(result_dir, "*.mp4"))  # 결과 폴더 내 .mp4 파일 리스트

for file in output_files:
    files.download(file)                # 결과 동영상을 내 컴퓨터로 다운로드
```
