âœ… 1. ì°¨ëŸ‰ì´ ì›€ì§ì´ëŠ” ì‹œë®¬ë ˆì´ì…˜ (ê¸°ë³¸ ì´ë™)

```python
# ì¼ì • ì‹œê°„ ë™ì•ˆ ìœ„ì¹˜ë¥¼ ì—…ë°ì´íŠ¸í•˜ëŠ” ì½”ë“œ (ê°€ì†ë„ëŠ” ê³ ë ¤í•˜ì§€ ì•Šê³  ë‹¨ìˆœ ì´ë™)
import math

dt = 1.0  # ì‹œê°„ ê°„ê²© (ì´ˆ)
x = 0.0   # ì´ˆê¸° x ì¢Œí‘œ
y = 0.0   # ì´ˆê¸° y ì¢Œí‘œ
velocity = 5.0 # ì†ë„ (ë‹¨ìœ„: m/s)
heading = math.pi / 4 # ì´ë™ ë°©í–¥ (ë¼ë””ì•ˆ, 45ë„)

x += velocity * dt * math.cos(heading)
y += velocity * dt * math.sin(heading)

print(f"ì°¨ëŸ‰ ìœ„ì¹˜: ({x:.2f}, {y:.2f})")
```

âœ… 2. ì†ë„ ì œì–´ ì‹œë®¬ë ˆì´ì…˜ (ëª©í‘œ ì†ë„ë¥¼ í–¥í•œ ê°€ì†/ê°ì†)

```python
# ë‹¨ìˆœí•œ Pì œì–´ (ë¹„ë¡€ì œì–´)
Kp = 0.5  # ë¹„ë¡€ ìƒìˆ˜
dt = 1.0  # ì‹œê°„ ê°„ê²©

# Define target_speed and max_speed
target_speed = 10.0 #ëª©í‘œë¡œ í•˜ëŠ” ì†ë„
max_speed = 15.0 #ì°¨ëŸ‰ì´ ë‚¼ ìˆ˜ ìˆëŠ” ìµœëŒ€ ì†ë„ ì œí•œ

# ì†ë„ ì˜¤ì°¨ ê³„ì‚°
speed_error = target_speed - velocity #velocity: í˜„ì¬ ì†ë„

# ìŠ¤ë¡œí‹€ ì œì–´ (ì˜¤ì°¨ì— ë¹„ë¡€)
throttle = Kp * speed_error #ì†ë„ì˜¤ì°¨ì— ë¹„ë¡€ìƒìˆ˜ë¥¼ ê³±í•´ì„œ ìŠ¤ë¡œí‹€ í¬ê¸° ì •í•¨
throttle = max(0.0, min(throttle, 1.0))  # ê°’ì„ 0~1 ì‚¬ì´ë¡œ ì œí•œ

# ì†ë„ ì—…ë°ì´íŠ¸ (ê°„ë‹¨ ëª¨ë¸: ê°€ì†ë„ = ìŠ¤ë¡œí‹€ * 3)
acceleration = throttle * 3.0
velocity += acceleration * dt #1ì´ˆ ë™ì•ˆ ê°€ì†ë„ë§Œí¼ ì†ë„ê°€ ëŠ˜ì–´ë‚œë‹¤ëŠ” ëœ»
velocity = min(velocity, max_speed) #ì†ë„ê°€ ìµœëŒ€ ì†ë„(max_speed)ë¥¼ ë„˜ì§€ ì•Šë„ë¡ ì œí•œí•¨.

print(f"í˜„ì¬ ì†ë„: {velocity:.2f} m/s, ìŠ¤ë¡œí‹€: {throttle:.2f}")
```

```python
waypoints = [[127.123, 37.456], [127.130, 37.460], [127.140, 37.470]]

def is_near_goal(current_pos, goal_pos, threshold=0.005):
    """
    í˜„ì¬ ìœ„ì¹˜(current_pos)ì™€ ëª©í‘œ ìœ„ì¹˜(goal_pos)ì˜ ê±°ë¦¬(ê²½ë„, ìœ„ë„ ì°¨ì´)ë¥¼ ê³„ì‚°í•´ì„œ,
    ë‘ ì ì´ threshold(ê¸°ë³¸ 0.005ë„) ì´ë‚´ë©´ True, ì•„ë‹ˆë©´ False ë°˜í™˜
    """
    lon_diff = abs(current_pos[0] - goal_pos[0]) #í˜„ì¬ ìœ„ì¹˜ì™€ ëª©í‘œ ìœ„ì¹˜ì˜ ê²½ë„ ì°¨ì´(lon_diff) ì™€ ìœ„ë„ ì°¨ì´(lat_diff) ë¥¼ ê°ê° ê³„ì‚°
    lat_diff = abs(current_pos[1] - goal_pos[1]) #abs() í•¨ìˆ˜ëŠ” ì ˆëŒ“ê°’, ì¦‰ ì°¨ì´ì˜ ë¶€í˜¸ë¥¼ ì—†ì• ê³  'ì–¼ë§ˆë‚˜ ë–¨ì–´ì ¸ ìˆëŠ”ì§€' í¬ê¸°ë§Œ ê³„ì‚°í•´ ì¤Œ.
    if lon_diff < threshold and lat_diff < threshold:
        return True
    else:
        return False

current_pos = waypoints[0]  # ì¶œë°œ ìœ„ì¹˜
goal_pos = waypoints[1]     # ë‹¤ìŒ ëª©í‘œ ìœ„ì¹˜

if is_near_goal(current_pos, goal_pos): #is_near_goal í•¨ìˆ˜ë¥¼ í˜¸ì¶œí•´ì„œ í˜„ì¬ ìœ„ì¹˜ì™€ ëª©í‘œ ìœ„ì¹˜ê°€ ê°€ê¹Œìš´ì§€ ê²€ì‚¬.
    print("ëª©í‘œ ìœ„ì¹˜ì— ê±°ì˜ ë„ì°©í–ˆì–´ìš”!")
else:
    print("ì•„ì§ ëª©í‘œ ìœ„ì¹˜ê¹Œì§€ ê±°ë¦¬ê°€ ë‚¨ì•˜ì–´ìš”.")
```
